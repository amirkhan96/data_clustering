```python
# last edition Amir Mahmood rafiey
import pandas as pd
from pandas import read_excel
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
```

```python
def optimal_clusters(data_frame):
    
    number_clusters = range(2,11)
    
    x_vector_list = list()
    for i in number_clusters:
        x_vector_list.append(i)
        
    inertia_list = list()
    
    change_percent = list()
    
    for number_cluster in number_clusters:
        model_fit = KMeans(n_clusters=number_cluster).fit(employee_data_1995_clustering)
        cluster_list = model_fit.predict(employee_data_1995_clustering)
        inertia_list.append(model_fit.inertia_)
        
    plt.plot(x_vector_list, inertia_list, 'bx-')
    plt.ylabel('sum of distances to cluster center')
    plt.title('The Elbow Method')
    plt.show()
    
    for i in range(1, len(x_vector_list)):
        change_percent.append((inertia_list[i] - inertia_list[i - 1])/ inertia_list[i - 1] * 100)

    return change_percent
```

```python
def draw_plot(x , y , color_colomn,title):
    
    plt.figure(figsize=(15,9))
    plt.title(title)
    
    plt.scatter(x, y,c=color_colomn, s=50, cmap='viridis')
```

```python
employee_data = pd.read_excel('Question_2_original.xlsx',header=1)
employee_data.columns = ['year', 'Region/country', 'under_1_millions', 'under_1_percent', 'under_2_millions', 'under_2_percent']
employee_data = employee_data.dropna(how='all')
employee_data = employee_data.drop(employee_data.index[0])
employee_data = employee_data.drop(employee_data.index[4201])
employee_data = employee_data.drop(employee_data.index[4200])


employee_data
```

```python
empty_colomn = list()
for i in range(0, employee_data.shape[0]):
    empty_colomn.append(0)
    
len(empty_colomn)
```

```python
employee_data['empty_colomns'] = empty_colomn
employee_data
```

```python
employee_data_1995 = employee_data[employee_data['year']==1995]
employee_data_1995 = employee_data_1995.sort_values(by=['under_2_percent'])
employee_data_1995
```

```python
employee_data_2010 = employee_data[employee_data['year']==2010]
employee_data_2010 = employee_data_2010.sort_values(by=['under_2_percent'])
employee_data_2010
```

```python
employee_data_2010_clustering = employee_data_2010.filter(items= ['under_2_percent', 'empty_colomns'])
employee_data_2010_clustering
```

```python
employee_data_1995_clustering = employee_data_1995.filter(items= ['under_2_percent', 'empty_colomns'])
employee_data_1995_clustering
```

```python
optimal_clusters(employee_data_1995_clustering)
```

```python
optimal_clusters(employee_data_2010_clustering)
```

```python
optimized_number_cluster = 5
model_fit = KMeans(n_clusters=optimized_number_cluster).fit(employee_data_1995_clustering)
cluster_list = model_fit.predict(employee_data_1995_clustering)
employee_data_1995['cluster_number'] = cluster_list
employee_data_1995
```

```python
optimized_number_cluster = 5
model_fit = KMeans(n_clusters=optimized_number_cluster).fit(employee_data_2010_clustering)
cluster_list = model_fit.predict(employee_data_2010_clustering)

employee_data_2010['cluster_number'] = cluster_list
employee_data_2010
```

```python
draw_plot(employee_data_1995.loc[:, 'under_2_percent'] , employee_data_1995.loc[:, 'empty_colomns'] , 
          employee_data_1995.loc[:, 'cluster_number'],'Under two, percent, 1995')
```

```python
draw_plot(employee_data_1995.loc[:, 'under_2_millions'] , employee_data_1995.loc[:, 'empty_colomns'] , 
          employee_data_1995.loc[:, 'cluster_number'],'Under two, million, 1995')
```

```python
draw_plot(employee_data_2010.loc[:, 'under_2_percent'] , employee_data_2010.loc[:, 'empty_colomns'] , 
          employee_data_2010.loc[:, 'cluster_number'],'Under two, percent, 2010')
```

```python
draw_plot(employee_data_2010.loc[:, 'under_2_millions'] , employee_data_2010.loc[:, 'empty_colomns'] , 
          employee_data_2010.loc[:, 'cluster_number'],'Under two, million, 2010')
```
